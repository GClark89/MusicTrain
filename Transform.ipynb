```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

#Brining back the original csv to clean and prepal for the model
file_path = r"/Users/sa8/Desktop/MusicTrain/data/train.csv"

df = pd.read_csv(file_path)
```

```python
#Dropped columns with objects in them and dropped the unnamed column as well as it doesn't seem useful 
df = df.drop(columns=["lyrics","artist_name","track_name","topic","genre","Unnamed: 0" , "release_date"], axis= 1)

```

```python


# Lets scale the len column to perfect match the other columns in our dataset to prepare for k-means clustering
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit and transform the 'Len' column 
df['len_scaled'] = scaler.fit_transform(df[['len']])

df['len_scaled'].to_frame()
```

```python
#Now since we have len scaled we can drop the original len column 
df = df.drop(columns=['len'])
```
